```python
import streamlit as st
import pandas as pd
from sklearn.cluster import KMeans  # Assuming you have KMeans trained

# Emoji dictionary for cluster personalities (replace with your choices)
personality_emoji = {
    0: " (Outgoing)",
    1: " (Analytical)",
    2: " (Sensitive)",
    3: " (Determined)",
    4: " (Creative)"
}

# Title and Introduction with CSS for styling
st.title("**K-Means Clustering Prediction App**")
st.markdown("""
<style>
body {
  background-color: #f0f0f5;
}
</style>
""", unsafe_allow_html=True)

st.write("Upload your data and discover personality clusters!")

# User Input with color customization
data_file = st.file_uploader("Choose a CSV file", type="csv", key="data-upload")
st.markdown("""
<style>
.css-1jcdk7u {  /* Target the file uploader container */
  background-color: #e0e0e0;
  border-radius: 5px;
  padding: 10px;
}
</style>
""", unsafe_allow_html=True)

if data_file is not None:
    df_new = pd.read_csv(data_file)

    # KMeans prediction (replace with your specific model)
    kmeans = KMeans(n_clusters=5)  # Assuming 5 clusters
    predictions = kmeans.predict(df_new)
    df_new['Cluster'] = predictions

    # Display data with predictions and color customization
    st.dataframe(df_new, style="table-striped")
    st.markdown("""
    <style>
    .stElement thead tr th {  /* Target table header */
      background-color: #d0d0d0;
      color: black;
    }
    </style>
    """, unsafe_allow_html=True)

    # Interactive Cluster Exploration with expanders and emoji display
    for cluster in df_new['Cluster'].unique():
        with st.expander(f"**Cluster {cluster}**"):
            cluster_data = df_new[df_new['Cluster'] == cluster]
            st.write(f"**Cluster Summary:** (Replace with actual summary)")
            st.write(personality_emoji[cluster])

```
