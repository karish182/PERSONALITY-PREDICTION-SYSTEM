VISUALISATION OF QUENTIONS AND THEIR DISTRIBUTION

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Function to plot the distribution of a personality trait
def plot_trait_distribution(columns, questions, trait_name):
    plt.figure(figsize=(12, 6))
    for i, column in enumerate(columns):
        plt.subplot(2, 5, i + 1)
        sns.countplot(data=df, x=column)
        plt.title(questions[column], fontsize=10)
        plt.xticks(rotation=45)
    plt.tight_layout()
    plt.suptitle(f"Distribution of {trait_name} Questions", fontsize=16)
    plt.show()
```

```python
# Plot distributions for each personality trait
plot_trait_distribution(EXT, ext_questions, 'Extraversion')
```

```python
plot_trait_distribution(EST, est_questions, 'Emotional Stability')
```

```python
plot_trait_distribution(AGR, agr_questions, 'Agreeableness')
```

```python
plot_trait_distribution(CSN, csn_questions, 'Conscientiousness')
```

```python
plot_trait_distribution(OPN, opn_questions, 'Openness')
```
